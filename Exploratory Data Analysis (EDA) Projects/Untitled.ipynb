Data Cleaning

```python
#Checking Missing Values

df.isnull().sum()
```

```python
#Handling Missing Values

df['Saving accounts'].fillna('N/A', inplace=True)
df['Checking account'].fillna('N/A', inplace=True)
```

```python
#Checking Duplicates Values

df.duplicated().sum()
```

Univariate Analysis

```python
#Age Distribution

#Histogram
sns.histplot(df['Age'], bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
#Credit Amount

sns.histplot(df["Credit amount"], bins=40)
plt.title("Credit Amount Distribution")
plt.show()
```

```python
#Loan Duaration

sns.countplot(data=df, x="Duration")
plt.title("Loan Duration")
plt.show()
```

```python
#Gender

sns.countplot(x=df["Sex"])
plt.title("Gender Distribution")
plt.show()
```

```python
#Housing

sns.countplot(x=df["Housing"])
plt.title("Housing Status")
plt.show()
```

```python
#Saving Accounts

sns.countplot(x=df["Saving accounts"])
plt.title("Saving Account Status")
plt.show()
```

Bivariate Analysis (Risk Relationships)

```python
#Age vs Credit Amount

sns.scatterplot(x="Age", y="Credit amount", data=df)
plt.title("Age vs Credit Amount")
plt.show()
```

```python
#Duration vs Credit Amount

sns.scatterplot(x="Duration", y="Credit amount", data=df)
plt.title("Duration vs Credit Amount")
plt.show()
```

```python
#Purpose vs Credit Amount

plt.figure(figsize=(10,5))
sns.barplot(x="Purpose", y="Credit amount", data=df)
plt.xticks(rotation=45)
plt.title("Purpose vs Credit Amount")
plt.show()
```
